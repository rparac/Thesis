Parses ILASP training output files to get the results

```python
import math
import os
import re

import numpy as np

base_dir = '/home/rp218/luke-for-roko/logs/generalisation/10-thesis-experiment-runs'
# file_extension = '.out'
file_extension = '.txt'

files = []
for file in os.listdir(base_dir):
    if file.endswith(file_extension):
        files.append(file)

files
```

```python

matches = []
for file in files:
    files_txt = open(f'{base_dir}/{file}').read()
    matches.extend(re.findall(r".*jaccard:(\d+\.\d+)/precision:(\d+\.\d+)/recall:(\d+\.\d+)", files_txt))
    print(matches)

matches = [(float(a), float(b), float(c)) for a, b, c in matches]
matches
```

```python
jaccards, precisions, recalls = zip(*matches)
jaccards = np.array(jaccards)
precisions = np.array(precisions)
recalls = np.array(recalls)
```

```python
avg = np.mean(jaccards)
std = np.std(jaccards)
se = std / math.sqrt(jaccards.shape[0])

print(f"Jaccard: mean {avg}, std {std}, se {se} ")
```

```python
avg = np.mean(precisions)
std = np.std(precisions)
se = std / math.sqrt(precisions.shape[0])

print(f"Precision: mean {avg}, std {std}, se {se} ")
```

```python
avg = np.mean(recalls)
std = np.std(recalls)
se = std / math.sqrt(recalls.shape[0])

print(f"Recall: mean {avg}, std {std}, se {se} ")
```
